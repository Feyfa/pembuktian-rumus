```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
```

```python
sns.set_style('whitegrid')
```

```python
plt.figure(figsize=(8,5))
plt.plot(x,y)
plt.title('test plot sinus sederhana')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
```
