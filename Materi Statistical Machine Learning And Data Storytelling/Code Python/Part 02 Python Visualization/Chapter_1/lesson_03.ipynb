# Perbedaan / Overview matplotlib dan seaborn

## Matplotlib
- tujuan: kontrol penuh, flexsibel
- level: low-level
- input data: array numpy, list
- sintaks: lebih panjang
- aturan praktis: gunakan untuk kostomisasi mendalam

## Seaborn
- tujuan: analisis statistika & estetik
- level: high-level
- input data: pandas dataframe
- sintaks: lebih ringkas
- aturan praktis: gunakan untuk explorasi cepat

1. MATPLOTLIB

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    'pengeluaran': [10,15,12,18,22],
    'penjualan': [100,140,110,180,210]
}
df = pd.DataFrame(data)
print(df)
```

```python
# membuat plot
plt.figure(figsize=(8,5))
plt.scatter(df['pengeluaran'], df['penjualan'])
plt.title('hubungan iklan dan penjualan (matplotlib')
plt.xlabel('pengeluaran iklan (Rp. 1000)')
plt.ylabel('penjualan (Rp. 10.000)')
plt.grid(True)
plt.show()
```

2. SEABORN

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='pengeluaran',y='penjualan')
plt.title('hubungan iklan dan penjualan (seaborn)')
plt.xlabel('pengeluaran iklan (Rp. 1000)')
plt.ylabel('penjualan (Rp. 10.000)')
plt.show()
```
