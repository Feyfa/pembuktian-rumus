# Langkah 1: memuat dan mengenal data

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as spd
```

```python
# atur tema default untuk semua plot kita
sns.set_theme(style='whitegrid', context='talk')

# muat dataset titanic
titanic_df = sns.load_dataset('titanic')
print(titanic_df.head())
# print(titanic_df.info())
```

### Langkah 2: Babak 1 - The Setup (Menetapkan Masalah)
mulai dengan gambaran besar: berapa banyak yang selamat dan berapa banyak yang tidak selamat

```python
 # Visual 1: Proporsi Selamat Vs Tidak Selamat
ax = sns.countplot(data=titanic_df,x='survived')

# kustomisasi penjelasannya
ax.set_title('Hasil Akhir: Mayoritas Penumpang Tidak Selamat', fontsize=16)
ax.set_xticks([0, 1])
ax.set_xticklabels(['tidak selamat', 'selamat'])
ax.set_xlabel('')
ax.set_ylabel('jumlah penumpang')
plt.show()
```

Narasi awal: dari 891 penumpang dalam data ini, lebih dari 60% tidak selamat. Siapa yang termasuk kelompok yang selamat?

### Langkah 3: Babak 2 - The Confrontation (Menemukan Wawasan)
Membedah faktor-faktor penentu

#### Investigasi 1: apakah "wanita dan ana

```python
# visual 2: Tingkat selamat bedasarkan gender
sns.catplot(data=titanic_df,x='sex',y='survived',kind='bar',height=6)
plt.title('tingkat selamat bedasarkan gender')
plt.ylabel('proporsi selamat')
plt.xlabel('jenis kelamin')
plt.show()
```

```python
# visual 3: Tingkat selamat bedasarkan usia
sns.catplot(data=titanic_df,x='survived',y='age',kind='violin',height=6,split=True,hue='sex')
plt.title('distribusi usia selamat vs tidak selamat')
plt.xticks([0,1],['tidak selamat', 'selamat'])
plt.ylabel('usia')
plt.xlabel('')
plt.show()
```

Temuan:
- Jenis kelamin: menjadi seoarang wanita secara dramatis meningkatkan peluang
   selamat. hampir 75% wanita selamat, dibandingkan dengan 20% penumpang pria
- Usia: distribusi usia untuk yang selamat sedikit lebih muda, terutama terlihat dari tonjolan pada anak anak kecil. ada hipotesa didukung yaitu anaka-anak di dahulu kan

### Investigasi 2: Apakah uang bisa membeli keselamatan

```python
# visual 4: Tingkat selamat bedasarkan kelas penumpang
sns.catplot(data=titanic_df,x='pclass',y='survived',kind='bar',height=6)
plt.title('tingkat selamat bedasarkan kelas')
plt.xticks([0,1,2],['Kelas 1', 'Kelas 2', 'Kelas 3'])
plt.ylabel('proporsi selamat')
plt.xlabel('kelas penumpang')
plt.show()
```

temuan: kelas sosial sangat berpengaruh. lebih dari 60% penumpang kelas 1 selamat
