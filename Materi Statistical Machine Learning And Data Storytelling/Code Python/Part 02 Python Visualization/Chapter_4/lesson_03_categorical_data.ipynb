# Categorical Data

### 1. Barplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as spd
import numpy as np
```

```python
# data
tips_df = sns.load_dataset('tips')
sns.set_theme(style='whitegrid')
print(tips_df.head())
```

```python
# membuat barplot
plt.figure(figsize=(8,4))
sns.barplot(
    data=tips_df,
    x='day',
    y='total_bill'
)
plt.title('Rata Rata Tagihan Per Hari')
plt.xlabel('Hari')
plt.ylabel('rata-rata tagihan, ratus (rp)')
plt.show()
```

```python
# membuat barplot
plt.figure(figsize=(8,4))
sns.barplot(
    data=tips_df,
    x='day',
    y='total_bill',
    estimator=np.median,
    errorbar=None
)
plt.title('Median Tagihan Per Hari')
plt.xlabel('Hari')
plt.ylabel('median tagihan, ratus (rp)')
plt.show()
```

### 2. countplot

```python
# membuat countplot
plt.figure(figsize=(8,4))
sns.countplot(
    data=tips_df,
    x='day',
    hue='time'
)

plt.title('jumlah pelangggan per hari')
plt.xlabel('hari')
plt.ylabel('jumlah')
plt.show()
```
