# visualizing relationship

### 1. Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as spd
import numpy as np
```

```python
tips_df = sns.load_dataset('tips')
print(tips_df.head())
```

```python
# membuat scatter plot
sns.set_theme(style='whitegrid')
plt.figure(figsize=(8,4))
sns.scatterplot(
    tips_df,
    x='total_bill',
    y='tip',
    hue='sex',
    style='smoker'
)
plt.title('Hubungan Total Tagihan Dan Tip')
plt.show()
```

### 2. Lineplot

```python
# dataset 'fmri'
fmri_df = sns.load_dataset('fmri')
print(fmri_df)
```

```python
# membuat lineplot
plt.figure(figsize=(8,4))
sns.lineplot(
    data=fmri_df,
    x='timepoint',
    y='signal',
    hue='event'
)

plt.figure(figsize=(8,4))
sns.scatterplot(
    data=fmri_df,
    x='timepoint',
    y='signal',
    hue='event'
)

plt.title('Perkembangan Sinyal Otak Dari Waktu Ke Waktu')
plt.show()
```

### 3. relplot

```python
plt.figure(figsize=(8,4))
sns.relplot(
    data=tips_df,
    x='total_bill',
    y='tip',
    col='day'
)
plt.show()
```
