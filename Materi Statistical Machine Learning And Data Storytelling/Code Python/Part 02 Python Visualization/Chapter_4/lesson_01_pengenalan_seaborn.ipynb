### 1. membuat plot dengan seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# membuat dataset dari bawaan seaborn
tips_df = sns.load_dataset('tips')
print(tips_df.head())
```

```python
sns.set_theme(
    style='whitegrid'
)
sns.scatterplot(
    data=tips_df,
    x='total_bill',
    y='tip'
)

plt.show()
```

```python
# contoh dengan matplotlib
plt.figure(
    figsize=(6.5,4)
)
plt.scatter(
    tips_df['total_bill'],
    tips_df['tip']
);
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```
