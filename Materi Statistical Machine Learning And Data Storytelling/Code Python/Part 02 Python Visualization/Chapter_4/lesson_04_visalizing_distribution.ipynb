### 1. boxplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# membuat data
tips_df = sns.load_dataset('tips')
sns.set_theme(style='whitegrid')
print(tips_df)
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=tips_df,x='day',y='total_bill',hue='sex')
plt.title('distibusi total tagihan per hari')
plt.show();
```

### 2. violint plot

```python
plt.figure(figsize=(8,4))
sns.violinplot(data=tips_df,x='day',y='total_bill',hue='sex')
plt.title('distibusi total tagihan per hari')
plt.show();
```

#### 3. strip plot

```python
plt.figure(figsize=(8,4))
sns.stripplot(data=tips_df,x='smoker',y='tip',jitter=True)
plt.title('sebaran perokok vs non-perokok')
plt.show();
```
