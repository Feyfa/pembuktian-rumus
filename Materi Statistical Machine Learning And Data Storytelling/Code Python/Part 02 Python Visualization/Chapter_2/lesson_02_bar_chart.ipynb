## Bar Chart

```python
# langkah 1: import matplotlib
import matplotlib.pyplot as plt
```

```python
# langkah 2: Data
# Sumbu X: merk laptop
produk = ['apple','acer','hp','lenovo','dell','asus']
# sumbu y: jumlah terjual
terjual = [10,20,5,30,5,60]
```

```python
# langkah ke 3: bar chart
plt.figure(figsize=(4,3)) # ini adalah ukuran canvas nya, berarti 400 x 300
plt.bar(
    produk, # sumbu x
    terjual, # sumbu y
    color='r', # warna merah
) # untuk menampilkan bar chart nya
plt.show()
```

```python
# langkah ke 3: bar chart multiple warna
plt.figure(figsize=(4,3)) # ini adalah ukuran canvas nya, berarti 400 x 300
warna = ['#663399','#aadd88','#ea12ae','#def123','#a321a3','#fed876'] # list warna
plt.bar(
    produk, # sumbu x
    terjual, # sumbu y
    color=warna, # warna merah
    edgecolor='black', # untuk border color
) # untuk menampilkan bar chart nya
plt.show()
```

```python
# langkah ke 3: bar chart multiple warna horizontal
plt.figure(figsize=(4,3)) # ini adalah ukuran canvas nya, berarti 400 x 300
warna = ['#663399','#aadd88','#ea12ae','#def123','#a321a3','#fed876'] # list warna
plt.barh(
    produk, # sumbu x
    terjual, # sumbu y
    color=warna, # warna merah
    edgecolor='black', # untuk border color
) # untuk menampilkan bar chart nya
plt.show()
```
