# Scatter Plot

```python
# langkah 1: import library matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# langkah 2: membuat data
# sumbu x: pengeluaran iklan (dalam juta rupiah)
pengeluaran_iklan = [2.5, 3.0, 3.2, 4.5, 5.0, 5.5, 6.0, 6.2, 7.0, 8.0]
# sumbu y: penjualan dalam unit
penjualan = [150, 180, 190, 250, 280, 310, 340, 330, 380, 420]
```

```python
# langkah 3: membuat figure
plt.figure(figsize=(4,4))
plt.scatter(pengeluaran_iklan, penjualan)
plt.show()
```

```python
# langkah 4: menambahkan style
plt.figure(figsize=(4,4))
plt.scatter(
    pengeluaran_iklan, penjualan)
plt.show()
```

```python
# langkah 4:  menambahkan style
plt.figure(figsize=(4,4))
plt.scatter(
    pengeluaran_iklan, # sumbu x
    penjualan,  # sumbu y
    color='red', # color
    s=100, # size diameter
    alpha=0.8, # opacity color
    edgecolors='black', # border color
    marker='o', # bentuk dari titik koordinatnya
)
plt.show()
```

```python
penjualan = [150, 180, 190, 250, 280, 310, 340, 330, 380, 420]
warna = 1 + 0.1 * np.arange(len(penjualan))

# langkah 5:  membuat gradasi
plt.figure(figsize=(4,4))
plt.scatter(
    pengeluaran_iklan, # sumbu x
    penjualan,  # sumbu y
    c=warna, # color
    cmap='viridis',
    s=100, # size diameter
    alpha=0.8, # opacity color
    edgecolors='black', # border color
    marker='o', # bentuk dari titik koordinatnya
)

# menghitung regresi
# menggunakan polyfit untuk regresi, derajatnya 1 linear
beta_1, beta_0 = np.polyfit(pengeluaran_iklan, penjualan, 1)
print({'beta_1': beta_1, 'beta_0': beta_0})

# membuat garis
plt.plot(
    np.array(pengeluaran_iklan),  # sumbu x
    beta_1 * np.array(pengeluaran_iklan) + beta_0, # sumbu y
    color='r', # color merah
    linestyle='--', # format garis nya
    linewidth=2, # untuk ketebalan garis
);

plt.show()
```
