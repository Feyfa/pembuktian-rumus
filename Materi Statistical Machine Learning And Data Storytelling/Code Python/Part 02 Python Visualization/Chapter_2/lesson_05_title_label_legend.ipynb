# Kustomisasi title, label, legend

```python
import matplotlib.pyplot as plt
import numpy as np

# data dasar
bulan = ['jan','feb','mar','apr','mei','jun']
penjualan_A = [200,220,250,240,280,300]
penjualan_B = [180,190,210,230,250,260]

plt.figure(figsize=(8,4))
plt.plot(
    bulan, # sumbu x
    penjualan_A, # sumbu x
    marker='o', # bentuk dari titik x,y
    label='Gadget A', # label
)
plt.plot(
    bulan, # sumbu x
    penjualan_B, # sumbu x
    marker='s', # bentuk dari titik x,y
    label='Gadget B', # label
)
plt.show()
```

```python
# 2. menyesuaikan judul
plt.figure(figsize=(8,4))
plt.plot(
    bulan, # sumbu x
    penjualan_A, # sumbu x
    marker='o', # bentuk dari titik x,y
    label='Gadget A', # label
)
plt.plot(
    bulan, # sumbu x
    penjualan_B, # sumbu x
    marker='s', # bentuk dari titik x,y
    label='Gadget B', # label
)
plt.title(
    'perbandingan penjualan Gadget A dan Gadget B', # ini title nya
    fontsize=12,
    fontweight='bold',
    color='darkblue',
    loc='left',
)
plt.show()
```

```python
# 3. menambah label
plt.figure(figsize=(8,4))
plt.plot(
    bulan, # sumbu x
    penjualan_A, # sumbu x
    marker='o', # bentuk dari titik x,y
    label='Gadget A', # label
)
plt.plot(
    bulan, # sumbu x
    penjualan_B, # sumbu x
    marker='s', # bentuk dari titik x,y
    label='Gadget B', # label
)
plt.title(
    'perbandingan penjualan Gadget A dan Gadget B', # ini title nya
    fontsize=12,
    fontweight='bold',
    color='darkblue',
    loc='left',
)
plt.xlabel('bulan')
plt.ylabel('unit terjual')
plt.show()
```

```python
# 4. custom label
plt.figure(figsize=(8,4))
plt.plot(
    bulan, # sumbu x
    penjualan_A, # sumbu x
    marker='o', # bentuk dari titik x,y
    label='Gadget A', # label
)
plt.plot(
    bulan, # sumbu x
    penjualan_B, # sumbu x
    marker='s', # bentuk dari titik x,y
    label='Gadget B', # label
)
plt.title(
    'perbandingan penjualan Gadget A dan Gadget B', # ini title nya
    fontsize=12,
    fontweight='bold',
    color='darkblue',
    loc='left',
)
plt.xlabel(
    'bulan',
    fontsize=12,
    color='red'
)
plt.ylabel(
    'unit terjual',
    fontsize=12,
    color='brown'
)
plt.show()
```

```python
# 5. menambahkan legend
plt.figure(figsize=(8,4))
plt.plot(
    bulan, # sumbu x
    penjualan_A, # sumbu x
    marker='o', # bentuk dari titik x,y
    label='Gadget A', # label
)
plt.plot(
    bulan, # sumbu x
    penjualan_B, # sumbu x
    marker='s', # bentuk dari titik x,y
    label='Gadget B', # label
)
plt.title(
    'perbandingan penjualan Gadget A dan Gadget B', # ini title nya
    fontsize=12,
    fontweight='bold',
    color='darkblue',
    loc='left',
)
plt.xlabel(
    'bulan',
    fontsize=12,
    color='red'
)
plt.ylabel(
    'unit terjual',
    fontsize=12,
    color='brown'
)
plt.legend()
plt.show()
```

```python
# 6. kustom legend
plt.figure(figsize=(8,4))
plt.plot(
    bulan, # sumbu x
    penjualan_A, # sumbu x
    marker='o', # bentuk dari titik x,y
    label='Gadget A', # label
)
plt.plot(
    bulan, # sumbu x
    penjualan_B, # sumbu x
    marker='s', # bentuk dari titik x,y
    label='Gadget B', # label
)
plt.title(
    'perbandingan penjualan Gadget A dan Gadget B', # ini title nya
    fontsize=12,
    fontweight='bold',
    color='darkblue',
    loc='left',
)
plt.xlabel(
    'bulan',
    fontsize=12,
    color='red'
)
plt.ylabel(
    'unit terjual',
    fontsize=12,
    color='brown'
)
plt.legend(
    loc='lower right',
    ncol=2,
    title='produk',
    shadow=True,
)
plt.grid(
    linestyle='--', 
    alpha=0.5
)
plt.show()
```
