untuk membuat garis grafik, kita bisa gunakan plt.plot()

1. Grafik paling dasar

```python
# Langkah 1: Import library matplotlib
import matplotlib.pyplot as plt
```

```python
# Langkah 2: Siapkan data kita
# sumbu X: bulan
bulan = ['Jan','Feb','Mar','Apr','Mei','Jun']
# sumbu Y: suhu dalam celcius
suhu_celcius = [27,28,29,31,30,29]
```

```python
# Langkah 3: membuat plot
plt.plot(
    bulan, # sumbu x
    suhu_celcius, # sumbu y
)
plt.show()
```

```python
# langkah 4: buat plot dengan warna berbeda
plt.plot(
    bulan, # sumbu x
    suhu_celcius, # sumbu y
    color='r' # untuk warna
)
plt.show()
```

```python
# langkah 5: buat plot dengan warna berbeda dan style yang berbeda
plt.plot(
    bulan, # sumbu x
    suhu_celcius, # sumbu y
    color='r', # untuk warna
    linestyle=':'
)
plt.show()
```

```python
# langkah 6: buat plot dengan warna berbeda dan style yang berbeda dan marker
plt.plot(
    bulan, # sumbu x
    suhu_celcius, # sumbu y
    color='r', # untuk warna
    linestyle='--', # garis merah akan menjadi -
    marker='o', # untuk memberikan titik merah pada (x,y)
)
plt.show()
```

```python
# contoh penggunaan line chart

# sumbu X: bulan
bulan = ['Jan','Feb','Mar','Apr','Mei','Jun']
# sumbu Y: suhu dalam celcius
suhu_jakarta = [27,28,29,31,30,29]
suhu_bandung = [23,24,25,27,26,25]
```

```python
plt.plot(
    bulan, # sumbu x
    suhu_jakarta, # sumbu y
    marker='o', # titik pada (x,y)
    color='r', # warna merah
    linestyle='-',
)
plt.plot(
    bulan, # sumbu x
    suhu_bandung, # sumbu y
    marker='o', # titik pada (x,y)
    color='b', # warna biru
    linestyle='--',
)
plt.show()
```
