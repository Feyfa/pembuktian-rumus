```python
import matplotlib.pyplot as plt
import numpy as np

# data dasar
bulan = ['jan','feb','mar','apr','mei','jun']
penjualan_A = [200,220,250,240,280,300]
penjualan_B = [180,190,210,230,250,260]

# 6. kustom legend
plt.figure(figsize=(8,4))
plt.plot(
    bulan, # sumbu x
    penjualan_A, # sumbu x
    marker='o', # bentuk dari titik x,y
    label='Gadget A', # label
)
plt.plot(
    bulan, # sumbu x
    penjualan_B, # sumbu x
    marker='s', # bentuk dari titik x,y
    label='Gadget B', # label
)
plt.title(
    'perbandingan penjualan Gadget A dan Gadget B', # ini title nya
    fontsize=12,
    fontweight='bold',
    color='darkblue',
    loc='left',
)
plt.xlabel(
    'bulan',
    fontsize=12,
    color='red'
)
plt.ylabel(
    'unit terjual',
    fontsize=12,
    color='brown'
)
plt.legend(
    loc='lower right',
    ncol=2,
    title='produk',
    shadow=True,
)
plt.grid(
    linestyle='--', 
    alpha=0.5
)

# 1. saving plot (harus dilakukan sebelum plt.show())
plt.savefig('plot1.png', transparent=True) # raster
plt.savefig('plot2.jpg', dpi=100) # raster
plt.savefig('plot3.pdf') # vector
plt.savefig('plot4.svg') # vector
plt.savefig('plot5.jpg', dpi=100, bbox_inches='tight') # raster

plt.show()
```
