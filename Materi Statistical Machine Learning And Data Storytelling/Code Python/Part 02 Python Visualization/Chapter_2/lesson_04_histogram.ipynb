# HISTOGRAM

histogram adalah visualisasi yang penting memahami distribusi. menggunakan interval atau "bins" untuk membagi data kita dan memberikan frequncy pada setiap interval

"pada rentang usia berapa pelanggan kami"

### 1. histogram dasar

```python
# langkah ke 1
import matplotlib.pyplot as plt
import numpy as np
```

```python
# langkah ke 2: siapkan data untuk contoh
# membuat 100 nilai anak yang terdistribusi normal dengan rata" 75
# dan standar deviasi adalah 10

np.random.seed(42)
nilai_ujian = np.random.normal(
    loc=75, # ini adalah rata rata
    scale=10, # ini adalah standar deviasi
    size=100, # ini adalah berapa banyak data yang diminta
)
print(nilai_ujian);
```

```python
# langkah ke 3: membuat histogram nya
plt.figure(figsize=(6,6))
plt.hist(nilai_ujian)
plt.show()
```

### 2. mengontrol interval

```python
# langkah ke 3: membuat histogram nya
plt.figure(figsize=(6,6))
plt.hist(
    nilai_ujian,
    bins=5, # ini adalah total kotak yang bakal dipecah di histogram
)
plt.show()
```

### 3. mengubah style

```python
# langkah ke 3: membuat histogram nya
plt.figure(figsize=(6,6))
plt.hist(
    nilai_ujian,
    bins=12, # ini adalah total kotak yang bakal dipecah di histogram
    color='purple', # color purple
    edgecolor='black', # border color black
    alpha=0.7, # opacitiy nya 0.5
)
plt.grid(
    axis='y', # garis potong nya acuannya dari sumbu y
    linestyle='--', # style garis nya adalah --
    alpha=0.5, # opacitiy nya 0.5
)

# menambahkan garis vertikal untuk nilai rata-rata
plt.axvline(
    nilai_ujian.mean(), # nilai x nya dimana
    color='red', # warna merah
    linestyle=':', # style garis nya : 
    linewidth=2, # lebar garis nya 2px
)

plt.show()
```

### 4. membandingkan distribusi

```python
np.random.seed(0)
nilai_ujian_1 = np.random.normal(
    loc=75, # ini adalah rata rata
    scale=8, # ini adalah standar deviasi
    size=100, # ini adalah berapa banyak data yang diminta
)
nilai_ujian_2 = np.random.normal(
    loc=72, # ini adalah rata rata
    scale=12, # ini adalah standar deviasi
    size=100, # ini adalah berapa banyak data yang diminta
)

plt.figure(figsize=(6,6))

plt.hist(
    nilai_ujian_1,
    bins=12, # ini adalah total kotak yang bakal dipecah di histogram
    color='blue', # color purple
    edgecolor='black', # border color black
    alpha=0.5, # opacitiy nya 0.5
    label='Kelas A' # untuk label
)
# menambahkan garis vertikal untuk nilai rata-rata
plt.axvline(
    nilai_ujian_1.mean(), # nilai x nya dimana
    color='blue', # warna merah
    linestyle=':', # style garis nya : 
    linewidth=2, # lebar garis nya 2px
)

plt.hist(
    nilai_ujian_2,
    bins=12, # ini adalah total kotak yang bakal dipecah di histogram
    color='orange', # color purple
    edgecolor='black', # border color black
    alpha=0.5, # opacitiy nya 0.5
    label='Kelas B' # untuk label
)
# menambahkan garis vertikal untuk nilai rata-rata
plt.axvline(
    nilai_ujian_2.mean(), # nilai x nya dimana
    color='orange', # warna merah
    linestyle=':', # style garis nya : 
    linewidth=2, # lebar garis nya 2px
)

plt.grid(
    axis='y', # garis potong nya acuannya dari sumbu y
    linestyle='--', # style garis nya adalah --
    alpha=0.5, # opacitiy nya 0.5
)

plt.show()
```
