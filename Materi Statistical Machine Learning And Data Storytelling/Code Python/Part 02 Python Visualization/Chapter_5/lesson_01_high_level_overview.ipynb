### 1. pairplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# mengambil dataset penguins
penguins_df = sns.load_dataset('penguins')
print(penguins_df)
```

```python
sns.pairplot(penguins_df,hue='island')
plt.show()
```

### 2. joint plot

fokus pada hubungan dua variable

```python
tips_df = sns.load_dataset('tips')
sns.set_theme(style='whitegrid')
sns.jointplot(tips_df,x='total_bill',y='tip')
plt.show()
```

```python
sns.jointplot(tips_df,x='total_bill',y='tip',kind='reg')
plt.show()
```

```python
sns.jointplot(tips_df,x='total_bill',y='tip',kind='hex')
plt.show()
```
