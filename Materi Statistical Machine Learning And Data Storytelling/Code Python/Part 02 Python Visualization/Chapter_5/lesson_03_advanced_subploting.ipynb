### catplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips_df = sns.load_dataset('tips')
sns.set_theme(style='whitegrid')
print(tips_df)
```

```python
sns.catplot(
    data=tips_df,
    kind='box',
    x='day',
    y='total_bill',
    col='time'
);
plt.show();
```

### facetgrid

grid yang terstruktur: 
- data
- row
- col
- hue

```python
# langkah 1: inisialisasi grid
g = sns.FacetGrid(
    tips_df,
    row='time',
    col='sex',
    hue='smoker',
    margin_titles=True
);

# langkah 2: petakan fungsi scatterplot
g.map(
    sns.scatterplot,
    'total_bill',
    'tip',
)
plt.show()
```
