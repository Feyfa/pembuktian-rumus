## Studi Kasus: Memahami Faktor Harga Berlian
### Tujuan bisnis:
- Sebuha perusahaan perhiasan ingin memahami faktor faktor apa saja yang paling memengaruhi harga berlian. Mereka memberi kita dataset "diamond" untuk dianalisis.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
diamond_tf = sns.load_dataset("diamonds")
print(diamond_tf)
```

### Langkah 1: Gambaran umum dengan pairplot

```python
# mengambil sample acak 500 baris untuk pairplot
sample_df = diamond_tf.sample(n=500, random_state=42)
print(sample_df)

# membuat pairplot dengan hue bedasarkan kualitas potongan ('cut')
sns.pairplot(sample_df, hue='cut', palette='viridis')

plt.show()
```
