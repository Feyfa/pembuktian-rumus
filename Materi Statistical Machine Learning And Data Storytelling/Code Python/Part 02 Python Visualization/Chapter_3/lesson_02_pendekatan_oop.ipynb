### plotting pendekatan oop

```python
# sebelumnya

import matplotlib.pyplot as plt

bulan = ['jan','feb','mar','apr']
penjualan = [100,200,150,300]

plt.figure(figsize=(4,4))
plt.plot(
    bulan, # sumbu x
    penjualan, # sumbu y
)
plt.show()
```

```python
bulan = ['jan','feb','mar','apr']
penjualan = [100,200,150,300]

fig, ax = plt.subplots(figsize=(4,4))

fig.suptitle('grafik perusahaan')
fig.supxlabel(2025)
fig.supylabel('laptop')

ax.plot(
    bulan, # sumbu x
    penjualan, # sumbu y
    label='unit', # label
)
ax.set_title('penjualan')
ax.set_xlabel('bulan')
ax.set_ylabel('jumlah')

ax.legend()

plt.tight_layout() # mengatur ulang layout agar jarak antar subplot tidak saling menimpa
plt.show()
```
