### mengatur batas sumbu

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
bulan_angka = [1,2,3,4,5,6,7,8]
penjualan = [200,220,250,240,280,300,310,330]
```

```python
fig, ax = plt.subplots(
    figsize=(10,6)
)

ax.plot(
    bulan_angka, # sumbu x
    penjualan, # sumbu y
    marker='o'
)
ax.set_title('fokus pada penjualan kuartal ke dua')
ax.set_xlabel('bulan(angka)')
ax.set_ylabel('unit terjual')

# batas sumbu y
ax.set_ylim(220,320)

# batas sumbu x
ax.set_xlim(3,7)

ax.grid()

plt.show()
```

### menyesuaikan tiks

```python
bulan = ['jan','feb','mar','apr','mei','jun','jul','agus']

fig, ax = plt.subplots(
    figsize=(10,6)
)

ax.plot(
    bulan, # sumbu x
    penjualan, # sumbu y
    marker='o'
)
ax.set_title('penjualan dengan label bulan yang disesuaikan')

# mengatur posisi tiks agar sesuai
# ticks adalah elemen" yang muncul di sumbu x contohnya, jan | feb | mar | apr | mei | jun | jul | agus
ax.set_xticks(bulan)
ax.set_xticklabels(bulan, rotation=45, ha='right')

plt.show()
```

### menambah anotasi dan text

```python
fig, ax = plt.subplots(
    figsize=(10,6)
)

ax.plot(
    bulan, # sumbu x
    penjualan, # sumbu y
    marker='o'
)
ax.set_title('fokus pada penjualan kuartal ke dua')
ax.set_xlabel('bulan(angka)')
ax.set_ylabel('unit terjual')

# anotasi, ini semacam text di tengah" koordinat untuk memberi peringatan
ax.annotate(
    'penurunan tak terduga', # title
    xy=('apr',240), # posisi x dan y
    xytext=('feb', 270),
    arrowprops=dict(
        facecolor='black',
        arrowstyle='->',
        connectionstyle='arc3,rad=-0.1'
    ), # untuk style panah yang menunjuk ke arah koordinat
    fontsize=12
)

ax.grid()

plt.show()
```

### anotasi persamaan

```python
bulan_angka = np.array([1,2,3,4,5,6,7,8])
penjualan = np.array([200,220,250,240,280,300,310,330])

fig, ax = plt.subplots(
    figsize=(8,4)
)

ax.scatter(
    bulan_angka, # sumbu x
    penjualan, # sumbu y
    marker='o'
)
ax.set_title('fokus pada penjualan kuartal ke dua')
ax.set_xlabel('bulan(angka)')
ax.set_ylabel('unit terjual')

# menghitung regresi
# menggunakan polyfit untuk regresi, derajatnya 1 linear
beta_1, beta_0 = np.polyfit(bulan_angka, penjualan, 1)

plt.plot(
    bulan_angka, # sumbu x
    beta_1 * bulan_angka + beta_0, # sumbu y
    color='red'
)

persamaan_text_1 = f'y = {beta_1:.2f}x + {beta_0:.2f}' # format biasa
persamaan_text_2 = r"$ \hat{y} $" + f"= {beta_1:.2f}x + {beta_0:.2f}" # format luar biasa
persamaan_text = f'{persamaan_text_1}   atau   {persamaan_text_2}'
ax.text(
    1,
    320, 
    persamaan_text, 
    fontsize=12, 
    bbox=dict(
        facecolor='white',
        alpha=0.5
    )
)

plt.show()
```
