### subplot

```python
# Langkah 1: import library
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Langkah 2: Data
bulan = ['jan','feb','mar','apr','mei','jun']
suhu = [27,28,29,31,30,29]
curah_hujan = [300,250,200,150,100,80]
```

```python
# langkah 3: membuat plot
plt.figure(figsize=(10,5))

# plot pertama
plt.subplot(
    2, # row
    1, # col
    1, # index
)
plt.plot(
    bulan, # data x
    suhu, # data y
)
plt.title('suhu rata rata bulanan');
plt.ylabel('C')

# plot kedua
plt.subplot(
    2, # row
    1, # col
    2, # index
)
plt.plot(
    bulan, # data x
    curah_hujan, # data y
)
plt.title('curah hujan rata rata bulanan');
plt.ylabel('mm')

plt.tight_layout() # mengatur ulang layout agar jarak antar subplot tidak saling menimpa
plt.show()
```

### subplots

```python
# subplots

fig, ax = plt.subplots(
    nrows=2,
    ncols=1,
    figsize=(8,8),
);

# ambil axis pertama
ax[0].plot(
    bulan, # sumbu x 
    suhu, # sumbu y
)
ax[0].set_title('suhu rata rata bulanan')

# ambil axis kedua
ax[1].plot(
    bulan, # sumbu x
    curah_hujan, # sumbu y
)
ax[1].set_title('curah hujan rata rata bulanan')
```

```python
# subplots

fig, ax = plt.subplots(
    nrows=2, # ukuran grid row, sebesar 2
    ncols=2, # ukuran grid col, sebesar 2
    figsize=(8,8), # ukuran figure
    # sharex=True, # untuk menyamakan scaling di sumbu x
    # sharey=True, # untuk menyamakan scaling di sumbu y
);

# ambil axis pertama
ax[0,0].plot(
    bulan, # sumbu x 
    suhu, # sumbu y
)
ax[0,0].set_title('suhu rata rata bulanan')

# ambil axis kedua
ax[0,1].plot(
    bulan, # sumbu x
    curah_hujan, # sumbu y
)
ax[0,1].set_title('curah hujan rata rata bulanan')

# data baru penjualan
penjualan = [200,150,200,100,160,180]
kepuasan_pelanggan = np.random.randint(70,100,6)

# ambil axis ketiga
ax[1,0].bar(
    bulan, # sumbu x
    penjualan, # sumbu y
)

# ambil axis keempat
ax[1,1].scatter(
    bulan, # sumbu x
    kepuasan_pelanggan, # sumbu y
)

# judul subplot utama
fig.suptitle("Dashboard Bulanan")
fig.tight_layout() # mengatur ulang layout agar jarak antar subplot tidak saling menimpa

plt.show()
```
