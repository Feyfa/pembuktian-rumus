style dan tema

```python
import matplotlib.pyplot as plt
import numpy as np
print(plt.style.available)
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,y1,label='sinx')
plt.plot(x,y2,label='cosx')
plt.title('plot style default matplotlib')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.legend()
plt.show()
```

GGPLOT

```python
# ngambil 
plt.style.use('ggplot')

plt.figure(figsize=(8,4))
plt.plot(x,y1,label='sinx')
plt.plot(x,y2,label='cosx')
plt.title('plot style default matplotlib')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.legend()
plt.show()
```

```python
# ngambil 
plt.style.use('dark_background')

plt.figure(figsize=(8,4))
plt.plot(x,y1,label='sinx')
plt.plot(x,y2,label='cosx')
plt.title('plot style default matplotlib')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.legend()
plt.show()
```
