```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# membuat data sampel
np.random.seed(10)

# np.random.normal(mean, std, jumlah_data)
pengeluaran_jakarta = np.random.normal(100,20,200);
pengeluaran_bandung = np.random.normal(80,30,200);
pengeluaran_surabaya = np.random.normal(90,25,200);
```

```python
# menggabungkan data ke dalam sebuah list
data = [pengeluaran_jakarta, pengeluaran_bandung, pengeluaran_surabaya]
# print(data)
```

```python
# figure dan axis
fig, ax = plt.subplots(
    figsize=(8,5)
)

# membuat boxplot
bp = ax.boxplot(
    data,
    patch_artist=True,
    notch=True
)

colors = ['lightblue','lightpink','lightgreen']
# bp['boxes'][0].set_facecolor('lightblue')
# bp['boxes'][1].set_facecolor('lightgreen')
# bp['boxes'][2].set_facecolor('lightpink')

for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

plt.show()
```

### Pie Chart

```python
# Data
sistem_operasi = ['android','ios','kaios','harmonyos']
pangsa_pasar = [50,25,15,10]
```

```python
# figure dan axis
fig, ax = plt.subplots(
    figsize=(8,5)
)

explode = (0, 0.1, 0, 0)

ax.pie(
    pangsa_pasar,
    labels=sistem_operasi,
    autopct='%1.1f%%',
    explode=explode,
    startangle=120
)

plt.show()
```
